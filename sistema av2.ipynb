```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import sqlite3


# Banco de dados
conexao = sqlite3.connect('sistema_estoque3.db')

c = conexao.cursor()

c.execute('''CREATE TABLE IF NOT EXISTS usuarios(
    nome text,
    senha_usuario text,
    email text,
    cpf text,
    endereco text
    )''')

c.execute('''CREATE TABLE IF NOT EXISTS itens_estoque(
    nome text,
    quantidade int
    )''')

conexao.commit()
conexao.close()

# Cores
azul = '#174dab'
rosa_bebe = '#fedbd8'
cor_cupcake = '#ca8b5f'
cinza = '#949494'

# Janela de login
tela_login = tk.Tk()
tela_login.geometry("400x350")
tela_login.title('Login')
tela_login.configure(bg=rosa_bebe)
tela_login.iconbitmap('bolinho.ico')
tela_login.resizable(width='false', height='false')

# Parte superior
frame_cima = tk.Frame(tela_login, width=400, height=100, bg=rosa_bebe)
frame_cima.grid(row=0, column=0, pady=1, padx=0, sticky='NSEW')

login = tk.Label(frame_cima, text='ATELIÊ DOCE TENTAÇÂO', font=(("Times", "24")), bg=rosa_bebe, fg=cinza)
login.place(x=5, y=25)

linha = tk.Label(frame_cima, text='', width=375, font=(("Times", "1")), bg=azul, fg=azul)
linha.place(x=10, y=70)

# Parte inferior
frame_baixo = tk.Frame(tela_login, width=400, height=250, bg=rosa_bebe)
frame_baixo.grid(row=1, column=0, pady=1, padx=0, sticky='NSEW')

nome = tk.Label(frame_baixo, text='NOME', font=(("Times", "18")), bg=rosa_bebe, fg=cinza)
nome.place(x=12, y=10)
e_nome = tk.Entry(frame_baixo, width=20, justify='left', font=(("", "15")), relief='solid', fg=cinza, bg='white')
e_nome.place(x=120, y=12)

senha = tk.Label(frame_baixo, text='SENHA', font=(("Times", "18")), bg=rosa_bebe, fg=cinza)
senha.place(x=12, y=100)
e_senha = tk.Entry(frame_baixo, width=20, justify='left', font=(("", "15")), relief='solid', fg=cinza, bg='white')
e_senha.place(x=120, y=100)




def tela_cadastro():
    tela_cadastro = tk.Tk()
    tela_cadastro.geometry("400x350")
    tela_cadastro.title('Cadastro Usuário')
    tela_cadastro.configure(bg=rosa_bebe)
    tela_cadastro.iconbitmap('bolinho.ico')
    tela_cadastro.resizable(width='false', height='false')

    # Parte superior
    frame_cima = tk.Frame(tela_cadastro, width=400, height=100, bg=rosa_bebe)
    frame_cima.grid(row=0, column=0, pady=1, padx=0, sticky='NSEW')

    b_voltar = tk.Button(frame_cima, text='Voltar', font=(("Times", "13")), bg=azul, fg='white',command=tela_cadastro.destroy)
    b_voltar.place(x=0, y=0)

    # Parte inferior
    frame_baixo = tk.Frame(tela_cadastro, width=400, height=250, bg=rosa_bebe)
    frame_baixo.grid(row=1, column=0, pady=1, padx=0, sticky='NSEW')

    nome = tk.Label(frame_baixo, text='NOME', font=(("Times", "18")), bg=rosa_bebe, fg=cinza)
    nome.place(x=12, y=5)
    e_nome = tk.Entry(frame_baixo, width=20, justify='left', font=(("", "15")), relief='solid', fg=cinza, bg='white')
    e_nome.place(x=120, y=7)

    senha = tk.Label(frame_baixo, text='SENHA', font=(("Times", "18")), bg=rosa_bebe, fg=cinza)
    senha.place(x=12, y=40)
    e_senha = tk.Entry(frame_baixo, width=20, justify='left', font=(("", "15")), relief='solid', fg=cinza, bg='white')
    e_senha.place(x=120, y=40)

    email = tk.Label(frame_baixo, text='EMAIL', font=(("Times", "18")), bg=rosa_bebe, fg=cinza)
    email.place(x=12, y=70)
    e_email = tk.Entry(frame_baixo, width=20, justify='left', font=(("", "15")), relief='solid', fg=cinza, bg='white')
    e_email.place(x=120, y=70)

    cpf = tk.Label(frame_baixo, text='CPF', font=(("Times", "18")), bg=rosa_bebe, fg=cinza)
    cpf.place(x=12, y=100)
    e_cpf = tk.Entry(frame_baixo, width=20, justify='left', font=(("", "15")), relief='solid', fg=cinza, bg='white')
    e_cpf.place(x=120, y=100)

    ende = tk.Label(frame_baixo, text='ENDEREÇO', font=(("Times", "14")), bg=rosa_bebe, fg=cinza)
    ende.place(x=12, y=130)
    e_ende = tk.Entry(frame_baixo, width=20, justify='left', font=(("", "15")), relief='solid', fg=cinza, bg='white')
    e_ende.place(x=120, y=130)

    def cadastrar(): #regras nome: maior que 2 senha: maior que 7 email: conter @ cpf: ter 11num end: opcicional
        nome = e_nome.get()
        senha = e_senha.get()
        email = e_email.get()
        cpf = e_cpf.get()
        endereco = e_ende.get()
        
        if len(nome) > 2 and len(senha) > 7 and '@' in email and len(cpf) == 11:
            conexao = sqlite3.connect('sistema_estoque3.db')
            c = conexao.cursor()
            c.execute('INSERT INTO usuarios VALUES(:nome, :senha_usuario, :email, :cpf, :endereco)',
                      {
                          'nome': nome,
                          'senha_usuario': senha,
                          'email': email,
                          'cpf': cpf,
                          'endereco': endereco
                      })
            e_nome.delete(0, 'end')
            e_senha.delete(0, 'end')  
            e_email.delete(0, 'end')
            e_cpf.delete(0, 'end')         
            e_ende.delete(0, 'end')         
        
            conexao.commit()
            conexao.close()


        else:
            messagebox.showinfo(title='erro', message='insira todas informaçoes corretamente')
            
    b_cadastro = tk.Button(frame_baixo, text='CADASTRAR', font=(("Times", "14")), bg=azul, fg='white', command=cadastrar)
    b_cadastro.place(x=150, y=180)        


b_cadastro = tk.Button(frame_baixo, text='CADASTRE-SE', font=(("Times", "14")), bg=azul, fg='white',command=tela_cadastro)
b_cadastro.place(x=120, y=150)

b_login = tk.Button(frame_baixo, text='LOGIN', font=(("Times", "20")), bg=azul, fg='white', command=lambda: login())
b_login.place(x=140, y=187)


def login():
    nome = e_nome.get()
    senha = e_senha.get()

    conexao = sqlite3.connect('sistema_estoque3.db')
    c = conexao.cursor()
    c.execute('SELECT * FROM usuarios WHERE nome = :nome AND senha_usuario = :senha', {'nome': nome, 'senha': senha})
    result = c.fetchone()
    if result:
        tela_login.destroy()
        tela_inicial()
        
    else:
        messagebox.showinfo(title='erro', message='Usuario ou senha nao encontrados')
        #mensagem_erro = tk.Label(frame_baixo, text='Usuário ou senha incorretos', font=(("", "12")), bg=rosa_bebe,fg='red')
        #mensagem_erro.place(x=90, y=127)
    conexao.commit()
    conexao.close()

        
def tela_inicial():
    tela_inicial = tk.Tk()
    tela_inicial.geometry("600x500")
    tela_inicial.title('Menu Inicial')
    tela_inicial.configure(bg=rosa_bebe)
    tela_inicial.iconbitmap('bolinho.ico')
    tela_inicial.resizable(width='false', height='false')
    
    def adicionar_banco():
        nome_produto = e_nome_produto.get()
        quantidade_produto = e_quantidade_produto.get()
        if len(nome_produto) > 1 and len(quantidade_produto) >= 1:
            conexao = sqlite3.connect('sistema_estoque3.db')
            c = conexao.cursor()
            c.execute('INSERT INTO itens_estoque VALUES(:nome, :quantidade)',{'nome': nome_produto, 'quantidade': quantidade_produto})         
            e_nome_produto.delete(0, 'end')
            e_quantidade_produto.delete(0, 'end')
            messagebox.showinfo(title='confirmaçao', message='Produto adicionado!')
            conexao.commit()
            conexao.close()
            tabela()
        else:
            messagebox.showinfo(title='erro', message='Preencha todos os campos!')

            
    def produto_no_banco():
        nome= e_nome_produto.get()
        conexao = sqlite3.connect('sistema_estoque3.db')
        c = conexao.cursor()
        c.execute('SELECT * FROM itens_estoque WHERE nome = :nome' , {'nome': nome }) 
        result = c.fetchone()
        if result:
            return False
        else:
            return True
        conexao.commit()
        conexao.close()
        tabela()
        

    def excluir_inf_banco():
        if produto_no_banco():
            messagebox.showinfo(title='erro', message='Produto nao encontrado!') 
            return
        conexao = sqlite3.connect('sistema_estoque3.db')
        c = conexao.cursor()
        c.execute(f'''
        DELETE FROM itens_estoque WHERE nome = '{e_nome_produto.get()}'
        
        ''')
        conexao.commit()
        conexao.close()
        tabela()
        messagebox.showinfo(title='confirmaçoa', message='Produto excluido!')
    
    
    
    def atualizar_banco():
        if len(e_quantidade_produto.get()) > 0  and produto_no_banco() == False:
            conexao = sqlite3.connect('sistema_estoque3.db')
            c = conexao.cursor()
            c.execute(f'''
            UPDATE  itens_estoque 
            SET quantidade = '{e_quantidade_produto.get()}'
            WHERE nome = '{e_nome_produto.get()}'

            ''')
            conexao.commit()
            conexao.close()
            tabela()
            messagebox.showinfo(title='confirmaçao', message='Produto atualizado!')
            
        else:
            messagebox.showinfo(title='erro', message='Preencha com nome e nova quantidade!') 
            return
        
    # Parte superior
    
    frame_cima = tk.Frame(tela_inicial, width=600, height=200, bg=rosa_bebe)
    frame_cima.grid(row=0, column=0, pady=1, padx=0, sticky='NSEW')
    
    quantidade_produto = tk.Label(frame_cima, text='QUANTIDADE', font=(("Times", "14")), bg=rosa_bebe,fg=cinza)
    quantidade_produto.place(x=110, y=80)
    e_quantidade_produto = tk.Entry(frame_cima, width=15, justify='left', font=(("", "15")), relief='solid',fg=cinza, bg='white')
    e_quantidade_produto.place(x=300, y=82)

    nome_produto = tk.Label(frame_cima, text='NOME', font=(("Times", "14")), bg=rosa_bebe,fg=cinza)
    nome_produto.place(x=110, y=40)
    e_nome_produto = tk.Entry(frame_cima, width=15, justify='left', font=(("", "15")), relief='solid',fg=cinza, bg='white')
    e_nome_produto.place(x=300, y=42)
    
    b_atualizar = tk.Button(frame_cima, text='ATUALIZAR', font=(("Times", "13")), bg=azul, fg='white',command=atualizar_banco)
    b_atualizar.place(x=100, y=160)
        
    b_excluir = tk.Button(frame_cima, text='EXCLUIR', font=(("Times", "13")), bg=azul, fg='white',command=excluir_inf_banco)
    b_excluir.place(x=240, y=160)

    b_adicionar = tk.Button(frame_cima, text='ADICIONAR', font=(("Times", "13")), bg=azul, fg='white',command=adicionar_banco)
    b_adicionar.place(x=380, y=160)
    
    #parte inferior
    
    frame_baixo = tk.Frame(tela_inicial, width=600, height=300, bg=rosa_bebe)
    frame_baixo.grid(row=1, column=0, pady=1, padx=0, sticky='NSEW')
    
    #tabela

    def tabela():
        
        conexao = sqlite3.connect('sistema_estoque3.db')
        c = conexao.cursor()
        c.execute("SELECT * FROM itens_estoque")
        items = c.fetchall()
        tree = ttk.Treeview(frame_baixo, columns=(1, 2), show="headings", height="50")
        tree.place(x=90, y=1)

        tree.heading(1, text='Produto')
        tree.heading(2, text='Quantidade')

        for item in items:
            tree.insert("", tk.END, values=item)

        conexao.commit()
        conexao.close()
        
    tabela()
        







tela_login.mainloop()
```
